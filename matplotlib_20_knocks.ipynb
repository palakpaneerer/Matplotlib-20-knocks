15.3D graph

```python
from mpl_toolkits.mplot3d import Axes3D

```

```python

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z= np.linspace(-2, 2, 100)
r = z **2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

# 3Dグラフ描く前の宣言
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(x, y, z, label='3D Curve')
ax.legend()
plt.show()


```

18.practice No.1 縦に2つグラフを並べる。

```python
# Data for the first graph
x1 = np.arange(10)
y1 = np.random.randint(-10, 10, 10)
```

```python
# Data for the second graph
x2 = ['Sam', 'Jonh', 'Kevin', 'Adam']
y2 = np.random.randint(0,200,4)
```

```python
# キャンバスのサイズを指定
fig = plt.figure(figsize=(9,6))

ax1 = fig.add_subplot(211)          #2行1列の1つ目
ax1.set_title('Result')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.plot(x1, y1)

ax2 = fig.add_subplot(212)
ax2.bar(x2, y2)

plt.show()

```

19.practice No.2 円グラフの一部を強調させる。

```python
labels = ['chips', 'hotdogs', 'cookies', 'chocolates']
sizes = [15, 30, 45, 10]
```

```python
explode = (0, 0.1, 0, 0)        # 円グラフから離して表示するパラメータ

plt.pie(sizes,explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')       # グラフの縦横の比を揃える。
plt.show()
```

20.practice No.3 積み上げタイプのヒストグラムを作成する。

```python
mu1, sigma1 = 100, 15
mu2, sigma2 = 90, 20
mu3, sigma3 = 110, 10
x1 = mu1 + sigma1 * np.random.randn(10000)
x2 = mu2 + sigma2 * np.random.randn(10000)
x3 = mu3 + sigma3 * np.random.randn(10000)
```

```python
colors = ['yellow', 'skyblue', 'blue']
labels = ['x1','x2','x3']

plt.hist([x1, x2, x3], bins=30, density=True, color=colors, label=labels, stacked=True)
plt.title('stacked histgram')
plt.xlabel('x')
plt.ylabel('freq')
plt.xlim(0,100)
plt.grid(True)
plt.legend(loc=2)
plt.show()
```
